##Questionamentos
- Informações do dataset:

```python
countrys = dataset['Country'].unique()
print(countrys)
```

```python
subregioes = ['Western Asia', 'Western Europe', 'Latin America and the Caribbean',
 'Eastern Europe', 'Northern Europe', 'Southern Europe', 'Eastern Asia',
 'Southern Asia', 'Central Asia', 'Northern America', 'Northern Africa',
 'Sub-Saharan Africa', 'Australia and New Zealand', 'Melanesia',
 'South-eastern Asia', 'Micronesia', 'Polynesia', 'Oceania', 'Asia', 'Europe', 'Americas', 'Africa', 'All Oceania']

dataset = dataset[~dataset['Country'].isin(subregioes)]
```

```python
# Encontrar o índice das linhas onde a coluna 'Country' tem o valor 'World'
indices_to_drop = dataset[dataset['Country'] == 'World'].index

# Usar drop para remover essas linhas
dataset = dataset.drop(indices_to_drop)
```

```python
# Encontrar o índice das linhas onde a coluna 'Region' tem o valor 'World'
indices_to_drop = dataset[dataset['Region'] == 'World'].index

# Usar drop para remover essas linhas
dataset = dataset.drop(indices_to_drop)
```

###Quais países apresentam as 10 maiores índices de homicídios nos últimos 5 anos?

```python
sub = dataset['Subregion'].unique()
print(sub)
```

```python
subregioes = ['Western Asia', 'Western Europe', 'Latin America and the Caribbean',
 'Eastern Europe', 'Northern Europe', 'Southern Europe', 'Eastern Asia',
 'Southern Asia', 'Central Asia', 'Northern America', 'Northern Africa',
 'Sub-Saharan Africa', 'Australia and New Zealand', 'Melanesia',
 'South-eastern Asia', 'Micronesia', 'Polynesia', 'Oceania', 'All Africa', 'All Americas', 'All Asia']

dataset = dataset[~dataset['Country'].isin(subregioes)]
```

```python
homicidios = dataset[dataset['Year'] > 2017]

homicidios = homicidios.groupby("Country")[["VALUE"]].sum().reset_index()
homicidios = homicidios.sort_values(by='VALUE', ascending=False).head(10)
```
